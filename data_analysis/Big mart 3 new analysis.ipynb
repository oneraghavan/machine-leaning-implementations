```python
def trasform_na_values(x):
    y = x.fillna(x.mean())
    return y    

train['Item_Weight'] = train.groupby("Item_Identifier").Item_Weight.transform(trasform_na_values)
```

```python
filtered_df = train[train.Item_Weight.notnull()]
# train['Item_Weight'] =  train['Item_Weight'].fillna(train['Item_Weight'].mean())
# filtered_df = train
for i in filtered_df['Item_Weight']:
    if np.isnan(i):
        print i
```

```python
df = pd.concat([filtered_df, pd.get_dummies(filtered_df['Outlet_Type'])], axis=1)
df = pd.concat([filtered_df, pd.get_dummies(filtered_df['Outlet_Location_Type'])], axis=1)
df = pd.concat([filtered_df, pd.get_dummies(filtered_df['Outlet_Identifier'])], axis=1)
df = pd.concat([filtered_df, pd.get_dummies(filtered_df['Outlet_Size'])], axis=1)
df = pd.concat([filtered_df, pd.get_dummies(filtered_df['Item_Type'])], axis=1)
df = pd.concat([filtered_df, pd.get_dummies(filtered_df['Item_Fat_Content'])], axis=1)
df = pd.concat([filtered_df, pd.get_dummies(filtered_df['Item_Identifier'])], axis=1)
del df['Outlet_Type']
del df['Outlet_Location_Type']
del df['Outlet_Identifier']
del df['Outlet_Size']
del df['Item_Type']
del df['Item_Fat_Content']
del df['Item_Identifier']
del df['Item_Outlet_Sales']
```
